```python
X = np.array(dataset)
## One-liner
neural_net = MLPRegressor(max_iter=2000).fit(X[:,:-1], X[:,-1])
## Result
res = neural_net.predict([[0.081304,0.04807,749.73]])
print(res)
```

```python

X = np.array(dataset)
X_train, X_test, y_train, y_test = train_test_split(X[:,:-1], X[:,-1],random_state=1)

## One-liner
neural_net = MLPRegressor(max_iter=2000).fit(X_train, y_train)
## Result
res = neural_net.predict(X_test)
print(res)
neural_net.score(X_test, y_test)
```

```python
import matplotlib.pyplot as plt

plt.plot(y_test, color='black', label='Actual Prices')
plt.plot(res, color='green', label="predicted prices")
plt.title("Predictions")
plt.xlabel('Time')
plt.ylabel('Price')
plt.legend(loc='upper left')
plt.show()
```
